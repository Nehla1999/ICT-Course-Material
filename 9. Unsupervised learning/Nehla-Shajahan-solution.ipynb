# Preparing the data

```python
wine1 = data.iloc[:, [0,12]].values
```

# Building the model - using elbow method to find the optimal no: of clusters

from sklearn.cluster import KMeans
wcss = []
for i in range(1,13):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state = 42)
    kmeans.fit(wine1)
    wcss.append(kmeans.inertia_)
    print(kmeans.inertia_)

```python
plt.plot(range(1,13),wcss)
plt.title("The elbow method")
plt.xlabel("no of clusters")
plt.ylabel("WCSS")
plt.show()
```

# Building the model with clusters = 3

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state = 42)
y_kmeans= kmeans.fit_predict(wine1)
```

```python
y_kmeans
```

# Hierarchial clustering

```python
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(wine1, method = "ward"))
plt.title('dendogram')
plt.xlabel('wine quality')
plt.ylabel('euclidean distance')
plt.show()
```

From the dendogram, we can see that then optimal no: of clusters = 2. 

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean',linkage = 'ward')
y_hc = hc.fit_predict(wine1)
```
